```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# sns.set('darkgrid')
```

```python
def read_csv_file(filename):
    df = pd.read_csv(filename, header=None, names=['x', 'y', 'p(x)'])
    return df
```

```python
kde_df = read_csv_file('output.csv')
```

```python
plot = sns.histplot(data=kde_df, x='y', kde=False)
plt.show()
```
