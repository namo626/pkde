```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# sns.set('darkgrid')
```

```python
# set seaborn style
sns.set(style="darkgrid")
```

```python
def read_csv_file(filename):
    df = pd.read_csv(filename, header=None, names=['x', 'y', 'p(x)'])
    return df
```

```python
kde_df = read_csv_file('output.csv')
```

```python
plt.figure(figsize=(12, 8))
plot = sns.histplot(data=kde_df, x='y', kde=False, stat='density', bins=100, color='blue', label='Histogram')
sns.rugplot(kde_df.x[::100], height=0.02, color='black',label='N = 8000 Samples')
sns.lineplot(data=kde_df, x='x', y='p(x)',color='red', label='KDE ')
plt.ylabel('p(x)')
plt.legend()
plt.show()
```
